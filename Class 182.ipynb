```python
df_copy=df.copy()
```

```python
df_copy.head()
```

```python
df_copy=df_copy.drop(df_copy.index[10472])
```

```python
df_copy[~df_copy['Reviews'].str.isnumeric()]
```

```python
df_copy['Reviews']=df_copy['Reviews'].astype(int)
```

```python
df_copy.info()
```

```python
df_copy['Size'].unique()
```

```python
df_copy['Size']=df_copy['Size'].str.replace('M','000')
df_copy['Size']=df_copy['Size'].str.replace('k','')
df_copy['Size']=df_copy['Size'].replace('Varies with device',np.nan)
df_copy['Size']=df_copy['Size'].astype(float)
```

```python
df_copy.info()
```

```python
df_copy['Installs'].unique()
```

```python
char_remove=['+',',','$']
cols_clean=['Installs','Price']
for item in char_remove:
    for col in cols_clean:
        df_copy[col]=df_copy[col].str.replace(item,'')
```

```python
df_copy['Installs'].unique()
```

```python
df_copy['Price'].unique()
```

```python
df_copy['Installs']=df_copy['Installs'].astype(int)
```

```python
df_copy['Price']=df_copy['Price'].astype(float)
```

```python
df_copy.info()
```

```python
df_copy['Last Updated'].unique()
```

```python
df_copy['Last Updated']=pd.to_datetime(df_copy['Last Updated'])
```

```python
df_copy['Day']=df_copy['Last Updated'].dt.day
df_copy['Month']=df_copy['Last Updated'].dt.month
df_copy['Year']=df_copy['Last Updated'].dt.year
```

# Class 182: EDA

```python
df_copy[df_copy.duplicated('App')].shape
```

```python
df_copy.drop_duplicates('App',keep='first',inplace=True)
```

```python
df_copy.shape
```

```python
num_ft = [ft for ft in df_copy.columns if df_copy[ft].dtype!='O']
cat_ft = [ft for ft in df_copy.columns if df_copy[ft].dtype=='O']
print('Num FT=',len(num_ft),num_ft)
print('Cat FT=',len(cat_ft),cat_ft)
```

```python
df_copy.info()
```

```python
df_copy.tail(1)
```

```python
for col in cat_ft:
    print(df_copy[col].value_counts(normalize=True)*100)
    print('____________________________________________')
```

```python
plt.figure(figsize=(15,15))
plt.suptitle("Univariate Analysis of num_ft", fontsize=20,fontweight='bold',alpha=0.8,y=1.)
for i in range(0,len(num_ft)):
    plt.subplot(5,3,i+1)
    sns.kdeplot(x=df_copy[num_ft[i]],shade=True,color='r')
    plt.xlabel(num_ft[i])
    plt.tight_layout()
```

```python
plt.figure(figsize=(15,15))
plt.suptitle("Univariate Analysis of cat_ft", fontsize=20,fontweight='bold',alpha=0.8,y=1.)
category=['Type','Content Rating']
for i in range(0,len(category)):
    plt.subplot(2,2,i+1)
    sns.countplot(x=df_copy[category[i]],palette='Set2')
    plt.xlabel(category[i])
    plt.xticks(rotation=45)
    plt.tight_layout()
```

```python
df_copy['Category'].value_counts().plot.pie(y=df_copy['Category'],figsize=(15,16),autopct='%1.1f')
```

```python
dfa= df_copy.groupby(['Category','App'])['Installs'].sum().reset_index()
dfa=dfa.sort_values('Installs',ascending=False)
apps = ['GAME', 'COMMUNICATION', 'PRODUCTIVITY', 'SOCIAL' ]
sns.set_context("poster")
sns.set_style("darkgrid")

plt.figure(figsize=(40,30))

for i,app in enumerate(apps):
    df2 = dfa[dfa.Category == app]
    df3 = df2.head(5)
    plt.subplot(4,2,i+1)
    sns.barplot(data= df3,x= 'Installs' ,y='App' )
    plt.xlabel('Installation in Millions')
    plt.ylabel('')
    plt.title(app,size = 20)

plt.tight_layout()
plt.subplots_adjust(hspace= .3)
plt.show()
```
